## Building a webscraping algorithm to get non-proprietary chemical names from compendium.ch


Compendium.ch contains the information on the non-proprietary name of our chemicals. Another reason that we will use this website is that many of the trademarks are for the german speaking market.

In webscraping, we  programatically extract information from a website, which is either in the HTML tags (simple case) or hidden in JavaScript (complicated). Before writing the code, it is worth checking the robots.txt file of the webpage to see which operations are allowed.
https://compendium.ch/robots.txt




In the case of compendium, all bots (User-agent: *) are allowed to crawl the entire website. To see a counter-example, have a look at comparis.ch.
https://www.comparis.ch/robots.txt


The following function can be used to retrieve the nonproprietary name or active substance of a given medicinal product. Since the text of the required HTML tag is hidden behind a JavaScript command, we need to use a selenium webdriver, which in contrast to a simple HTML request, can execute the JavaScript command. 
Some things to consider:

* We will iterate over a long list, therefore we have to make sure that the function can handle errors properly 
* When the webpage is loaded, we need to wait a bit to properly load the content, we use time.sleep() for this
* The function will try to retrieve the data 2 times, then will return False if unsuccessful


Useful links for webscraping: 
* HTML tags: https://www.w3schools.com/TAGS/default.asp
* XPath notation: https://devhints.io/xpath
* Python requests: https://requests.readthedocs.io/en/latest/
* Selenium: https://selenium-python.readthedocs.io/
* Book on webscraping: https://www.amazon.de/-/en/Ryan-Mitchell/dp/1491985577/ref=sr_1_3?crid=UWWQHDMQX5V3&keywords=Webschaber&qid=1672932087&sprefix=webscraping%2Caps%2C85&sr=8-3 

DM me for the PDF of the book ;)

```python
def get_nonprop_name(med, sleeptime = 1, patience = 0):
    
    if patience == 2: 
        return False
    
    url = f"https://compendium.ch/search?q={med}"
    fireFoxOptions = webdriver.FirefoxOptions()
    fireFoxOptions.headless = True
    driver = webdriver.Firefox(options=fireFoxOptions)
    driver.get(url)
    sleep(sleeptime)
    
    try:
        element = driver.find_element("xpath","//span[@class='no-color-link']")
        name = element.text
        if name == "+": 
            element = driver.find_elements("xpath","//span[@class='no-color-link']")[1]
            name = element.text
        driver.quit()
        return name

    except: 
        try: 
            element = driver.find_emelent("xpath", "//span[@class='product-out-of-trade']")
            name = element.text
            if name == "+": 
                element = driver.find_elements("xpath","//span[@class='product-out-of-trade']")[1]
                name = element.text
            driver.quit()
            return name
        except: 
            patience += 1
            sleeptime += 1
            text = get_nonprop_name(med, sleeptime, patience)
        
        return text
```

```python
# We use the function to get non-prop names of the medicines. We use 50 for demonstrational purposes, otherwise the whole procedure takes 5 hours..
nonprop_names = []
for i in tqdm(med_names[1200:1210]): 
    name = get_nonprop_name(i)
    nonprop_names.append(name)
```

## Extracting data about molecules from PubChem using the PUG-REST API

* With the following function, you can extract data from the PubChem database using a simple REST query, which is based on a GET request of a specific URL (documentation: https://pubchem.ncbi.nlm.nih.gov/docs/pug-rest). We will use the json format in the response object. 

* We use the f-string method in python to integrate the chemical name and search parameter into our query (https://peps.python.org/pep-0498/).

* We send a get request using the requests library. The status code of 200 in the response object indicates that the request was successful.

* We return the requested properties as a pandas DataFrame object.

```python
def get_properties(name, *args): 
    
    
    search_string = ",".join(args)
    
    url = f"https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/name/{name}/property/{search_string}/json"
    
    r = requests.get(url)
    if r.status_code == 200: 
        properties = r.json()["PropertyTable"]["Properties"][0]
        df = pd.DataFrame(properties, index = [0])
        return df
    else:
        return False

def generate_fp_matrix(df, fp): 
    
    arrays = []
    
    for fingerprint in df[fp]: 
        arrays.append([int(i) for i in fingerprint])
        
    return pd.DataFrame(np.array(arrays), index = df.Title)
```

## Fingerprint embedding of chemical formulas 

* To extract molecular features of chemical compounds or to calculate molecular similarity, we representate the formula in a numerically interpretable way. 

* There are a handful of fingerprints available with pros and cons (https://towardsdatascience.com/a-practical-introduction-to-the-use-of-molecular-fingerprints-in-drug-discovery-7f15021be2b1).

* In this excercise, we will gather the PubChem fingerprints, MACCS keys and Morgan fingerprints for the molecules

* Pubchem fingerprints are 881 bit embeddings, where each bit corresponds to the presence or absence of a given substructure in the molecule (https://ftp.ncbi.nlm.nih.gov/pubchem/specifications/pubchem_fingerprints.pdf).

* MACCS keys use 166 bit embeddings

* Morgan fingerprints are circular fingerprints




* Example: PubChem fingerprint representation






```python
df_main = pd.DataFrame()
for drug in tqdm(drug_names):
    df = get_properties(drug, "Title", "IUPACName","CanonicalSMILES","Fingerprint2D", "XLogP", "Charge", "HBondDonorCount", "HBondAcceptorCount", "MolecularWeight")
    #we will get a False (bool) if the chemical of the property is not in the PubChem database
    if type(df) != bool: 
        df_main = pd.concat([df_main, df])
```

```python
#Open the .csv file instead of the long operation 

df_main = pd.read_csv("molecules.csv", index_col = 0)
```

```python
 
#We decode from base64 to ASCII (https://en.wikipedia.org/wiki/ASCII)
df_main["FP_Pubchem"] = ["".join( ["{:08b}".format(x) for x in b64decode(fingerprint)] )[32:913] for fingerprint in df_main.Fingerprint2D]
# We can generate the MACCS keys from the SMILES string using the rdkit.Chem module
df_main["FP_MACCS"] = [MACCSkeys.GenMACCSKeys(Chem.MolFromSmiles(df_main.CanonicalSMILES.iloc[i])).ToBitString() for i in range(df_main.shape[0])]
# We can generate the MorganFP from the SMILES string using the rdkit.Chem module
df_main["FP_Morgan"] = [AllChem.GetMorganFingerprintAsBitVect(Chem.MolFromSmiles(df_main.CanonicalSMILES.iloc[i]), 5,nBits = 1024).ToBitString() for i in range(df_main.shape[0])]
```

## Calculating molecular similarity 

* Measuring molecular similarity can be applied for many cheminformatics tasks such as similarity searching, property prediction and molecular diversity analysis

* There is no definitive approach to assess similarity, but the two main componenets are: 
    - Molecular descriptors (fingerprints that we previously gathered)
    - Similarity metric ( a quantitative score for the degree of similarity)
    
* Tanimoto similarity metric (Jaccard index): 
    - Ratio of intersection over union 
    - In this context, the common on bits divided by the union of on bits 



```python
#Using the MACCS keys, generate a tanimoto similarity matrix 

def tanimoto_similarity(df): 
    compounds = df.index.values
    sim_matrix = []
    for i in tqdm(range(df.shape[0])):
        cmpd_x = df.iloc[i, :].values
        sims = []
        for j in range(df.shape[0]):
            cmpd_y = df.iloc[j, :].values
            common_bits  =  sum([all([cmpd_x[k] == cmpd_y[k] , cmpd_x[k] == 1])  for k in range(len(cmpd_x))])
            on_bits_i = sum([k == 1 for k in cmpd_x])
            on_bits_j = sum([k == 1 for k in cmpd_y])
            tan = common_bits/(on_bits_i + on_bits_j - common_bits)
            sims.append(tan)
        sim_matrix.append(sims)
    return pd.DataFrame(sim_matrix, index = compounds, columns = compounds)
```

## Gathering data on the side effects and therapeutic indications of drugs



```python
def get_mesh_ids(name): 
  url = f"http://ctdbase.org/tools/batchQuery.go?inputType=chem&inputTerms={name}&report=diseases_curated&format=json"
  r = requests.get(url)
  if r.status_code == 200: 
     json = r.json()
     try: 
         meshids_marker = [i["DiseaseID"].split(":")[1] for i in json if i["DirectEvidence"] == "marker/mechanism" and "DirectEvidence" in i.keys()] 
         meshids_thera = [i["DiseaseID"].split(":")[1] for i in json if i["DirectEvidence"] == "therapeutic" and "DirectEvidence" in i.keys()]
         disease_marker = [i["DiseaseCategories"].split(":")[0] for i in json if i["DirectEvidence"] == "marker/mechanism" and "DiseaseCategories" in i.keys()] 
         disease_thera = [i["DiseaseCategories"].split(":")[0] for i in json if i["DirectEvidence"] == "therapeutic" and "DiseaseCategories" in i.keys()]
         df = pd.DataFrame({"Title": name, "markerID": ",".join(meshids_marker), "theraID": ",".join(meshids_thera),"marker": ",".join(disease_marker), "thera": ",".join(disease_thera) }, index = [0])
         return df
     except:
         return pd.DataFrame({"Title": name, "markerID": False, "theraID": False, "marker": False, "thera": False}, index = [0])
```

```python
df_mesh = pd.DataFrame()
for drug in tqdm(df_main.Title.str.lower()):
    df = get_mesh_ids(drug)
    if type(df) != bool: 
        df_mesh = pd.concat([df_mesh, df])
```

```python
df_mesh["cardio_se"] = df_mesh.marker.str.lower().str.contains("cardiovascular")
df_mesh["neuro_se"] = df_mesh.marker.str.lower().str.contains("nervous system")
df_mesh["blood_se"] = df_mesh.marker.str.lower().str.contains("blood")
df_mesh["digestive_se"] = df_mesh.marker.str.lower().str.contains("digestive")
df_mesh["cancer_se"] = df_mesh.marker.str.lower().str.contains("cancer")

df_mesh["cardio_thera"] = df_mesh.thera.str.lower().str.contains("cardiovascular")
df_mesh["neuro_thera"] = df_mesh.thera.str.lower().str.contains("nervous system")
df_mesh["blood_thera"] = df_mesh.thera.str.lower().str.contains("blood")
df_mesh["digestive_thera"] = df_mesh.thera.str.lower().str.contains("digestive")
df_mesh["cancer_thera"] = df_mesh.thera.str.lower().str.contains("cancer")
```
